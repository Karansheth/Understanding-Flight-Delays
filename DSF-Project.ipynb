```python
columns = ['Station_Id', 'Date', 'Element_Type', 'Element_Value', 'Measurement_Flag', 'Quality_Flag', 'Source_Flag', 'Observation_Time']
weather_2018 = pd.read_csv('weather_2018.csv', names=columns)
weather_2018
```

```python
weather_code = ["USW00013874", "USW00014739", "USW00093721", "USW00013881", "USW00093738", "USW00003017", "USW00003927", "USW00094847", "USW00014734", "USW00012849", "USW00022521", "USW00093738", "USW00012960", "USW00094789", "USW00023169", 
            "USW00023174", "USW00014732", "USW00012815", "USW00014819", "USW00013893", "USW00012839", "USW00014922", "USW00094846", "USW00013739", "USW00023183", "USW00024229", "USW00024233", "USW00023234", "USW00024127", "USW00012842"]
```

```python
weather_busiest = weather_2018[weather_2018['Station_Id'].isin(weather_code)]
weather_busiest.reset_index(inplace=True)
weather_busiest.drop('index', axis=1, inplace=True)
weather_busiest
```

```python
def split_year(date):
    year = str(date)
    return str(year[0:4])
def split_month(date):
    month = str(date)
    return str(month[4:6])
def split_day(date):
    day = str(date)
    return str(day[6:8])
```

```python
weather_busiest['Year'] = weather_busiest['Date'].apply(lambda x:split_year(x))
weather_busiest['Month'] = weather_busiest['Date'].apply(lambda x:split_month(x))
weather_busiest['Day'] = weather_busiest['Date'].apply(lambda x:split_day(x))
weather_busiest
```

```python
weather_busiest.drop(['Date', 'Measurement_Flag', 'Quality_Flag', 'Source_Flag', 'Observation_Time'], axis=1, inplace=True)
weather_busiest
```

```python
weather_busiest['Element_Type'].unique()
```


1.   TMAX - Maximum Temperature
2.   TMIN - Minimum temperature
3.   PRCP - Precipitation
4.   SNOW - Snowfall (mm or inches)
5.   SNWD - Snow Depth (mm or inches)
6.   AWND - Average Dialy Wind Speed
7.   TAVG - Average Temperature
8.   PGTM -  Peak gust time (hours and minutes, i.e., HHMM)
9.   PSUN -  Daily percent of possible sunshine (percent)
10.  TSUN -   Daily total sunshine (minutes)
11.   WDF2 - Direction of fastest 2-minute wind
12.   WDF5 - Direction of fastest 5-second wind (degrees)
13.  WSF2 - Fastest 2-minute wind speed (miles per hour or meters per second as per user preference)
14.  WSF5 = Fastest 5-minute wind speed (miles per hour or meters per second as per user preference)
15.  WT01 - Fog, ice fog, or freezing fog (may include heavy fog)
16.  WT02 - Heavy fog or heaving freezing fog (not always distinguished from fog)
17.  WT03 - Thunder
18.  WT04 - Ice pellets, sleet, snow pellets, or small hail 
19.  WT05 - Hail (may include small hail)
20.  WT06 - Glaze or rime 
21.  WT07 - Dust, volcanic ash, blowing dust, blowing sand, or blowing obstruction
22.  WT08 - Smoke or haze
23.  WT09 - Blowing or drifting snow 
24.  WT10 - Tornado, waterspout, or funnel cloud
