```python
arr_del_class = []
for delay in df['ARR_DELAY_NEW']:
    if delay > 10:
        arr_del_class.append(1)
    else:
        arr_del_class.append(0) 
```

```python
df['ARR_DELAY_Class'] = arr_del_class
```

```python
df1 = df.copy()
```

```python
import xgboost as xgb
```

```python
X1 = df.drop('ARR_DELAY_NEW', axis=1)
y1 = df['ARR_DELAY_NEW']
```

```python
# Perform the split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3, random_state=42)
```

```python
xgb4 = xgb.XGBRegressor(objective='reg:squarederror', colsample_bytree = 0.4, learning_rate=0.1,
                        max_depth=5, alpha=10, n_estimators=400).fit(X_train1,y_train1)
xgb_pred4 = xgb4.predict(X_test1)                        
```

```python
rmspe4 = np.sqrt((np.mean(np.square(y_test1-xgb_pred4)/y_test1)))
rmspe4
```
