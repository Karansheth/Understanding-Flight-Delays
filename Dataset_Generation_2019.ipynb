```python
weather_busiest = weather_2019[weather_2019['Station_Id'].isin(weather_code)]
weather_busiest.reset_index(inplace=True)
weather_busiest.drop('index', axis=1, inplace=True)
weather_busiest
```

```python
def split_year(date):
    year = str(date)
    return str(year[0:4])
def split_month(date):
    month = str(date)
    return str(month[4:6])
def split_day(date):
    day = str(date)
    return str(day[6:8])
```

```python
weather_busiest['Year'] = weather_busiest['Date'].apply(lambda x:split_year(x))
weather_busiest['Month'] = weather_busiest['Date'].apply(lambda x:split_month(x))
weather_busiest['Day'] = weather_busiest['Date'].apply(lambda x:split_day(x))
weather_busiest
```

```python
weather_busiest.drop(['Date', 'Measurement_Flag', 'Quality_Flag', 'Source_Flag', 'Observation_Time'], axis=1, inplace=True)
weather_busiest
```


1.   TMAX - Maximum Temperature

17.  WT03 - Thunder
18.  WT04 - Ice pellets, sleet, snow pellets, or small hail 
19.  WT05 - Hail (may include small hail)
20.  WT06 - Glaze or rime 
21.  WT07 - Dust, volcanic ash, blowing dust, blowing sand, or blowing obstruction
22.  WT08 - Smoke or haze
23.  WT09 - Blowing or drifting snow 
24.  WT10 - Tornado, waterspout, or funnel cloud
